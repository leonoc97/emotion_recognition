```python
# Matplotlib config
%matplotlib inline

# SVG-Anzeige anschalten
%config InlineBackend.figure_formats = ['retina']

# Größe definieren
%config InlineBackend.rc = {'figure.figsize': (5.0, 3.0)}

import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv("../../data/Diamonds/diamonds.csv.bz2")

df = df.sample(100)

df.head()
```

Ein Jointplot setzt sich aus einen Scatterplot und Balkendiagramme zusammen und ermöglicht so eine übersichtliche Darstellung der 2 Diagrammarten. Dafür müssen auf der X- und auf der Y-Achse Zahlenwerte eingetragen werden (hier `carat` und `price`).

```python
sns.jointplot(x = df["carat"], y = df["price"]);
```

Die eingetragenen Punkte des Scatterplots lassen sich auch zur übersicht als Hex(agon) darstellen, je dunkler das Blau, desto mehr Werte liegen in diesen Hex.

```python
sns.jointplot(x = df["carat"], y = df["price"], kind = "hex");
```

Außerdem lässt sich die Größe (auf der X-Achse mit `xlim` und auf der Y-Achse mit `ylim`, bzw die gesamte Größe mit `gridsize`) und die Farbe mit (`color`) der einzelnen Hexagone verändern.

```python
sns.jointplot(x = df["carat"], 
              y = df["price"], 
              kind = "hex", 
              xlim = (0, 3), 
              ylim = (0, 15000), 
              gridsize = 15, 
              color = "k");
```
