```python
# Matplotlib config
%matplotlib inline

# SVG-Anzeige anschalten
%config InlineBackend.figure_formats = ['svg']

# Größe definieren
%config InlineBackend.rc = {'figure.figsize': (5.0, 3.0)}

import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv("../../data/Diamonds/diamonds.csv.bz2")

df.head()
```

Histogramme stellen die eingelesenen Werte als eine Gaußsche Verteilungskurve dar, wobei die Wahrscheinlichkeit, wie oft ein Wert vorkommt, errechnet werden kann indem man die X-Achsenwerte mit dem entsprechenden Y-Achsenwerten multipliziert.

```python
sns.histplot(x = df["depth"], binwidth = 1);
```

Zusätzlich dazu können wir auch einen Zahlenbereich angeben, für den die Bins generiert werden. Wichtig hierbei: Der letzte Bin startet, solange der Startwert vom Bin kleiner gleich dem Ende vom binrange ist. 

Das bedeutet hier (bei einer bindwidth von 2), der letzte Bin startet bei 63, und hört dann bei 65 auf (obwohl die 65 außerhalb vom binrange ist). 

Einträge, die außerhalb der Bins liegen, werden in dieser Grafik zudem komplett übersprungen.

```python
sns.histplot(x = df["depth"], binwidth = 2, binrange = (55, 64));
```
